```python
#아나콘다 프롬프트를 통해 folium 설치 후 사용
import folium
#파일 로드
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/corazzon/boostcourse-ds-510/master/data/medical_201909.csv.zip", low_memory=False)
df.shape
```

```python
#전국 노인/치매병원 데이터를 df_silver_hospital에 가져오기
df_silver_hospital = df[df["상권업종소분류명"]=="노인/치매병원"]
df_silver_hospital.tail(3)
```

```python
#geo_df에 df_silver_hospital 복사해서 할당
geo_df = df_silver_hospital.copy()
```

```python
#지도의 중심을 지정하기 위해 위도와 경도의 평균 구하기
fmap = folium.Map(location=[geo_df['위도'].mean(), geo_df['경도'].mean()], 
                 zoom_start=12)

for n in geo_df.index:
    #팝업에 들어갈 텍스트를 지정
    #상호명과 주소를 붙여서 출력
    popup_name = geo_df.loc[n, '상호명'] + ' - ' + geo_df.loc[n, '도로명주소']
    
    folium.Marker(
        location=[geo_df.loc[n,'위도'], geo_df.loc[n, '경도']],
        popup=popup_name, 
        tooltip=popup_name
    ).add_to(fmap)


fmap
```
