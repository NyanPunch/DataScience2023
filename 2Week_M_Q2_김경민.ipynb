```python
import matplotlib.pyplot as plt
from IPython.display import set_matplotlib_formats
plt.rc('font',family='Malgun Gothic')
plt.rc('axes', unicode_minus=False)

%config InlineBackend.figure_format = 'retina'

# 그래프가 노트북 안에 보이게 하기 위해
%matplotlib inline
```

```python
#파일 로드
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/corazzon/boostcourse-ds-510/master/data/medical_201909.csv.zip", low_memory=False)
df.shape
```

```python
import seaborn as sns
```

```python
df_animal = df[df["상권업종소분류명"] == "동물병원"]
df_animal.head(2)
```

```python
df_animal["시도명"].value_counts()
```

```python
#동물병원 비율로 환산하여 저장
v_animal = df_animal["시도명"].value_counts(normalize=True)
v_animal
```

```python
# 약국 데이터만 가져오기
df_medi = df[df["상권업종소분류명"] == "약국"]
df_medi.head(4)
#전국 도시별로 약국 수 비율로 저장
v_medi = df_medi["시도명"].value_counts(normalize=True)
v_medi
```

### 약국과 동물병원 비교

```python
#시도별 약국 파이 그래프
v_medi.plot.pie()
```

```python
#시도별 동물병원 파이 그래프
v_animal.plot.pie()
```

경기도와 서울의 그래프만 보아도 약국이 많은 곳에 동물병원이 많이 분포해있다는걸 알 수 있다.
